```python
def color_red_or_green(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

  if value < 0.1:
    color = 'red'
  elif ((value >= 0.1) and (value<0.3)):
    color = 'orange'
  elif ((value > 0.3) and (value <0.7)):
    color = 'black'  
  elif ((value >= 0.7) and (value<0.9)):
    color = 'blue'
  else:
    color = 'green'



  return 'color: %s' % color
```

```python
tr3=tr2.style.applymap(color_red_or_green)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Treasury Yield spreads.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
tr3.to_excel(writer, sheet_name='Sheet1')

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Sheet1']


# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
